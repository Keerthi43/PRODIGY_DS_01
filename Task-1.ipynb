```python
import pandas as pd
import matplotlib.pyplot as plt

data = r"API_SP.POP.TOTL_DS2_en_csv_v2_85.csv"

df = pd.read_csv(data, skiprows=4)

df = df.dropna(subset=['Country Name', '2022'])

df.sort_values(by='2022', ascending=False, inplace=True)

top_10 = df.head(10)

plt.figure(figsize=(15, 8))
plt.bar(top_10['Country Name'], top_10['2022'] / 1e9, color='black')
plt.xlabel('Country')
plt.ylabel('Population (Billions)')
plt.title('Top 10 Countries by Population in 2022')
plt.xticks(rotation=30,ha='right')

plt.tight_layout()
plt.show()
```
